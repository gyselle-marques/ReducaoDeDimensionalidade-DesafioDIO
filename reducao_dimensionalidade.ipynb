```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
#ler imagem, transformar em vetor numpy e exibir
imagem = mpimg.imread('./teste/chihiro.jpg')
imagem = np.array(imagem)
plt.imshow(imagem)
```

```python
#verifica formato da imagem
imagem.shape
```

```python
#converte a imagem para tons de cinza tilizando a expresso gray = 0.2989 * R + 0.5870 * G + 0.1140 * B
imagem_tons_de_cinza = 0.2989*imagem[:,:,0] + 0.5870*imagem[:,:,1] + 0.1140*imagem[:,:,2]

imagem_tons_de_cinza.shape
```

```python
plt.imshow(imagem_tons_de_cinza, cmap='gray')
```

```python
#converte a imagem em tons de cinza para binria
imagem_binaria = imagem_tons_de_cinza > 127
plt.imshow(imagem_binaria, cmap='gray')
```

```python
#exibi as trs imagens simultaneamente para comparao
plt.subplot(1, 3, 1)
plt.axis('off')
plt.imshow(imagem)
plt.subplot(1, 3, 2)
plt.axis('off')
plt.imshow(imagem_tons_de_cinza, cmap='gray')
plt.subplot(1, 3, 3)
plt.axis('off')
plt.imshow(imagem_binaria, cmap='gray')
plt.show()
```
